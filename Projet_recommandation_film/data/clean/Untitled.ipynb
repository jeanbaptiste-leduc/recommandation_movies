```python
#répartition par genre 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import ast

# Convertir en liste si c'est une chaîne
df['genres_list'] = df['genres_list'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)

# Une ligne par genre et enlever les espaces
df_exploded = df.explode('genres_list')
df_exploded['genres_list'] = df_exploded['genres_list'].str.strip()

# Compter les films par genre
genre_counts = df_exploded['genres_list'].value_counts().reset_index()
genre_counts.columns = ['Genre', 'Nombre']

couleurs = ['#7FA9F7', '#FFA3A6', '#FFE08A', '#70D8B0','#3D6BEB']
# Graphique
plt.figure(figsize=(14,8))
barplot = sns.barplot(x='Genre', y='Nombre', data=genre_counts, palette=couleurs)

# Ajout des nombres au-dessus des barres
for i, row in genre_counts.iterrows():
    barplot.text(i, row['Nombre'] + 0.5, row['Nombre'], ha='center', color='black')

plt.xticks(rotation=45)
plt.title('Répartition des films par genre', fontsize=20, fontweight='bold')
plt.ylabel('Nombre de films',fontweight='bold')
plt.xlabel('Genre',fontweight='bold')
plt.show()
```

```python
#5 plus importantes compagnies de production (camembert)

#Compter le nombre de compagnie de production
compagnie_counts = df['production_companies_name_list'].value_counts().reset_index()
compagnie_counts.columns = ['production_companies_name_list', 'Nombre']
print(compagnie_counts)
print(df['production_companies_name_list'].head(10))
print(df['production_companies_name_list'].apply(type).value_counts())
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# --- Étape 1 : garder uniquement les listes non vides ---
df_clean = df[df['production_companies_name_list'].apply(lambda x: len(x) > 0)]

# --- Étape 2 : extraire la première compagnie ---
df_clean['compagnie'] = df_clean['production_companies_name_list'].apply(lambda x: x[0])

# --- Étape 3 : compter le nombre de films par compagnie ---
compagnie_counts = df_clean['compagnie'].value_counts().reset_index()
compagnie_counts.columns = ['compagnie', 'Nombre']

# --- Étape 4 : top 5 ---
top5_compagnies = compagnie_counts.head(5)

# --- Étape 5 : palette de couleurs ---
couleurs = ['#7FA9F7', '#FFA3A6', '#FFE08A', '#70D8B0', '#3D6BEB']

# --- Étape 6 : créer le camembert ---
plt.figure(figsize=(8,8))
plt.pie(
    top5_compagnies['Nombre'],
    labels=top5_compagnies['compagnie'],
    autopct='%1.1f%%',
    startangle=140,
    colors=couleurs
)
plt.title("Répartition des films par compagnie (Top 5)")
plt.show()
```
